```python
# Load the dataset using pandas.
# Use matplotlib and seaborn for plotting.
import pandas as pd                  # Import pandas for data manipulation and analysis (DataFrames, CSV reading)
import matplotlib.pyplot as plt      # Import matplotlib pyplot for creating static visualizations
import seaborn as sns                # Import seaborn for statistical data visualization

df = pd.read_csv('iris.csv')         # Data Loading, Read the uploaded iris.csv file into a pandas DataFrame named 'df'
```

```python
# Print the shape, column names, and the first few rows using .head().
print("Shape:", df.shape)                          # Prints (rows, columns)
print("\nColumn Names:", df.columns.tolist())      # Prints list of column names
print("\nFirst few rows:")
print(df.head())                                   # Shows first 5 rows of the DataFrame to understand data structure
```

```python
# Use .info() for Dataset Info.
print("=== DATASET INFO ===")
df.info()           # Prints concise summary: column name, non-null counts, column data types, memory usage
```

```python
# Use describe() for summary statistics.
print("\n=== DESCRIPTIVE STATISTICS ===")
df.describe()          # Generates count, mean, std, min, 25%, 50%, 75%, max for numerical columns
```

```python
# Visualize Dataset: Create a scatter plot to show relationships between features.
# Create a 2x3 grid for 6 plots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))          # Creates 2 rows, 3 columns of subplots (6 total), figure size 15x10 inches.

# Plot 1: sepal_length vs sepal_width scatter plot
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', ax=axes[0,0])    # Creates colored scatter plot (different colors for each species) - shows relationship between sepal length and width
axes[0,0].set_title('Sepal Length vs Sepal Width')                                          # Adds title to first subplot

# Plot 2: sepal_length vs petal_length scatter plot
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species', ax=axes[0,1])   # Shows correlation between sepal length and petal length, colored by species
axes[0,1].set_title('Sepal Length vs Petal Length')                                         # Adds title to second subplot

# Plot 3: sepal_length vs petal_width scatter plot
sns.scatterplot(data=df, x='sepal_length', y='petal_width', hue='species', ax=axes[0,2])    # Shows how sepal length relates to petal width across different species
axes[0,2].set_title('Sepal Length vs Petal Width')                                          # Adds title to third subplot

# Plot 4: sepal_width vs petal_length scatter plot
sns.scatterplot(data=df, x='sepal_width', y='petal_length', hue='species', ax=axes[1,0])    # Visualizes relationship between sepal width and petal length across different species
axes[1,0].set_title('Sepal Width vs Petal Length')                                          # Adds title to forth subplot

# Plot 5: sepal_width vs petal_width scatter plot
sns.scatterplot(data=df, x='sepal_width', y='petal_width', hue='species', ax=axes[1,1])     # Shows how sepal width compares to petal width across different species
axes[1,1].set_title('Sepal Width vs Petal Width')                                           # Adds title to fifth subplot

# Plot 6: petal_length vs petal_width scatter plot
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', ax=axes[1,2])    # Shows petal length vs width clearly separates species
axes[1,2].set_title('Petal Length vs Petal Width')                                          # Adds title to sixth subplot

plt.tight_layout()       # Automatically adjusts subplot parameters to prevent overlap
plt.show()               # Displays all 6 scatter plots
```

```python
# Use histograms to show value distributions.
fig, axes = plt.subplots(2, 2, figsize=(12, 10))      # Creates 2x2 grid of subplots (4 total) size 12x10 inches

# Sepal Length Histogram
sns.histplot(df['sepal_length'], bins=20, kde=True, color='skyblue', ax=axes[0,0])   # Creates histogram with 20 bins, adds Kernel Density Estimation curve (smoothed distribution), skyblue color
axes[0,0].set_title('Sepal Length Distribution', fontsize=12)    # Sets title with font size 12
axes[0,0].set_xlabel('Sepal Length')      # Labels x-axis
axes[0,0].set_ylabel('Frequency')         # Labels y-axis

# Sepal Width Histogram
sns.histplot(df['sepal_width'], bins=20, kde=True, color='lightgreen', ax=axes[0,1])  # Histogram for sepal width distribution with light green color
axes[0,1].set_title('Sepal Width Distribution', fontsize=12)     # Sets title with font size 12
axes[0,1].set_xlabel('Sepal Width')      # Labels x-axis
axes[0,1].set_ylabel('Frequency')        # Labels y-axis

# Petal Length Histogram
sns.histplot(df['petal_length'], bins=20, kde=True, color='salmon', ax=axes[1,0])  # Shows distribution of petal lengths
axes[1,0].set_title('Petal Length Distribution', fontsize=12)      # Sets title with font size 12
axes[1,0].set_xlabel('Petal Length')    # Labels x-axis
axes[1,0].set_ylabel('Frequency')       # Labels y-axis

# Petal Width Histogram
sns.histplot(df['petal_width'], bins=20, kde=True, color='purple', ax=axes[1,1])    # Distribution of petal widths with purple color
axes[1,1].set_title('Petal Width Distribution', fontsize=12)      # Sets title with font size 12
axes[1,1].set_xlabel('Petal Width')      # Labels x-axis
axes[1,1].set_ylabel('Frequency')        # Labels y-axis

plt.tight_layout()       # Prevents label overlap between subplots
plt.show()               # Displays all 4 histograms
```

```python
# Use box plots to identify outliers.
fig, axes = plt.subplots(1, 4, figsize=(16, 6))  # Creates 1 row, 4 columns of box plots, figure size 16x6 inches

# Sepal Length Box Plot
sns.boxplot(y=df['sepal_length'], ax=axes[0], color='skyblue')   # Creates vertical box plot
axes[0].set_title('Sepal Length Box Plot', fontsize=12)       # Sets title with font size 12
axes[0].set_ylabel('Sepal Length')                            # Labels y-axis

# Sepal Width Box Plot
sns.boxplot(y=df['sepal_width'], ax=axes[1], color='lightgreen')   # Box plot for sepal width, shows outliers
axes[1].set_title('Sepal Width Box Plot', fontsize=12)       # Sets title with font size 12
axes[1].set_ylabel('Sepal Width')                            # Labels y-axis

# Petal Length Box Plot
sns.boxplot(y=df['petal_length'], ax=axes[2], color='salmon')    # Displays box plot of petal length
axes[2].set_title('Petal Length Box Plot', fontsize=12)      # Sets title with font size 12
axes[2].set_ylabel('Petal Length')                           # Labels y-axis

# Petal Width Box Plot
sns.boxplot(y=df['petal_width'], ax=axes[3], color='purple')    # Box plot for petal width
axes[3].set_title('Petal Width Box Plot', fontsize=12)       # Sets title with font size 12
axes[3].set_ylabel('Petal Width')                            # Labels y-axis

plt.tight_layout()    # Adjusts spacing between subplots
plt.show()            # Displays all 4 box plots

# Print outlier information (IQR Method)
print("\n=== OUTLIER DETECTION ===\n")

for column in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']: # Loop through each numerical feature
    Q1 = df[column].quantile(0.25)        # Calculates first quartile (25th percentile) for the column
    Q3 = df[column].quantile(0.75)        # Calculates third quartile (75th percentile)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR          # Calculates lower outlier threshold (1.5*IQR rule)
    upper_bound = Q3 + 1.5 * IQR          # Calculates upper outlier threshold
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]     # Filters rows where value is outside bounds

    print(f"{column}:")                   # Prints current column name
    print(f"  Lower bound: {lower_bound:.2f}")    # Prints lower threshold (2 decimal places)
    print(f"  Upper bound: {upper_bound:.2f}")    # Prints upper threshold
    print(f"  Number of outliers: {len(outliers)}")  # Count of outlier values found
    if len(outliers) > 0:      # If outliers exist
        print(f"  Outlier values: {outliers[column].tolist()}")   # Prints actual outlier values as list
    print()
```

```python
# Box Plots of Iris Features Grouped by Species.
fig, axes = plt.subplots(1, 4, figsize=(16, 5))  # Creates 1x4 grid for species-wise comparison

sns.boxplot(data=df, x='species', y='sepal_length', ax=axes[0])  # Compares sepal length distribution across three species (setosa, versicolor, virginica)
axes[0].set_title('Sepal Length by Species')

sns.boxplot(data=df, x='species', y='sepal_width', ax=axes[1])   # Shows sepal width variation by species - setosa typically has wider sepals
axes[1].set_title('Sepal Width by Species')

sns.boxplot(data=df, x='species', y='petal_length', ax=axes[2])  # Petal length comparison - setosa has smallest, virginica largest petal length
axes[2].set_title('Petal Length by Species')

sns.boxplot(data=df, x='species', y='petal_width', ax=axes[3])   # Petal width comparison - best discriminating feature between species
axes[3].set_title('Petal Width by Species')

plt.tight_layout()
plt.show()
```

```python
# Box Plots of Sepal and Petal Measurements (All Species Combined).
plt.figure(figsize=(10, 6))   # Creates new figure with size 10x6 inches
sns.boxplot(data=df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])  # Creates side-by-side box plots for all numerical features on same scale for easy comparison
plt.title('Box Plots for All Features', fontsize=14)  # Adds main title with larger font
plt.ylabel('Values', fontsize=12)  # Labels y-axis
plt.xticks(rotation=45)            # Rotates x-axis labels 45 degrees to prevent overlap
plt.show()
```
